```python
from collections import Counter

# 샘플 코퍼스
corpus = ["자연어 처리는 재미있다.", "자연어 처리는 컴퓨터 과학의 한 분야다.", "데이터 분석은 자연어 처리와 밀접하다."]

# 단어 단위로 코퍼스를 분리
tokenized_corpus = [sentence.split() for sentence in corpus]
tokenized_corpus
```

```python
flat_tokens = [word for sentence in tokenized_corpus for word in sentence]
flat_tokens
```

```python
# 단어 빈도 분석
word_freq = Counter(flat_tokens)
print("단어 빈도 분석 결과:", word_freq)
```

```python
from nltk.corpus import wordnet  # NTLK의 WordNet(영어 어휘/의미 사전) 코퍼스

# Synset 사용 예제
word = "bank"
synsets = wordnet.synsets(word)  # bank에 해당하는 synsets(의미 후보) 리스트
for syn in synsets:
    print(f"뜻: {syn.definition()}")  # 해당 synsets의 정의 출력
```
