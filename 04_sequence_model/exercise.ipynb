```python
import torch
import torch.nn as nn

# 간단한 RNN + Linear로 구성된 모델 (입력 시퀀스의 마지막 시점 정보를 사용해 출력 생성)
class SimpleRNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleRNN, self).__init__()                               # nn.Module 초기화
        self.rnn = nn.RNN(input_size, hidden_size, batch_first=True)    # RNN 레이어 (입력: (B:batch size, T:TimeSteps, F:입력벡터))
        self.fc = nn.Linear(hidden_size, output_size)                   # 마지막 은닉 상태 -> 출력 변환

    def forward(self, x):
        out, _ = self.rnn(x)            # RNN 순전파 (시점별 출력 out)
        out = self.fc(out[:, -1, :])    # 마지막 시점만 사용해 fc 통과
        return out
    
# 모델 초기화
model = SimpleRNN(input_size=10, hidden_size=20, output_size=1)
model
```
